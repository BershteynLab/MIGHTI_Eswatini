#Estimating Annual Probability of Death (p_death) and Duration of Condition (dur_condition) for Conditions in MIGHTI

## Purpose

This notebook guides users in estimating the annual conditional probability of death (p_death) and the average duration of disease (dur_condition) for health conditions, using publicly available data such as  Global Burden of Disease (GBD) data.


## Probability of death

The `p_death` parameter represents the **annual conditional probability of death** for an individual who has the disease of interest (e.g., T2D). This is distinct from general mortality metrics reported in sources like the Global Burden of Disease (GBD), which often give:

* Total number of deaths
* Deaths as a percent of all deaths
* Death rates per 100,000 population

These must be transformed into a **conditional risk of death given disease**.
\begin{align*}
    \text{deaths} &= \frac{\text{deaths_per_100k} * \text{population}}{100,000}\\
    \text{cases} &= \text{prevalence (%)} * \text{population}\\
    \text{p_death} &= \frac{\text{deaths}}{\text{cases}}= \frac{\text{deaths_per_100k}}{\text{prevalence (%)}}
\end{align*}

\begin{align*}
    \text{p_death} &= \frac{\text{deaths_per_100k}}{\text{prevalence (%)}}
\end{align*}


## Duration of conditions

The `dur_condition` parameter represents the **the estimated average duration of condition** in years, derived as:

$$ \text{dur_condition} = \frac{\text{Prevalence}}{\text{Incidence}}$$



## Example: Eswatini, T2D in 2007

* GBD T2D death rate: 55.81 per 100,000
* GBD T2D prevalence percent: 2.25%
* GBD T2D prevalence rate: 2171.05
* GBD T2D incidence rate: 177.33

We selected 2007 as the baseline year to estimate condition duration and mortality parameters (dur_condition, p_death) **for chronic diseases in Eswatini in the absence of intervention**, reflecting pre-expansion levels of healthcare access. This decision is based on the observation that Eswatini's health system began scaling up substantially after 2010, particularly due to the integration of HIV care infrastructure and broader investment in noncommunicable disease (NCD) services. The World Health Organization (WHO) noted significant improvements in service coverage, health workforce, and facility readiness after 2010, with increasing integration of chronic disease management into primary health care.

```python
# Basic Setup
def estimate_p_death(deaths_per_100k, prevalence_percent, population=1_000_000):

    if prevalence <= 0:
        raise ValueError("Prevalence must be greater than 0.")

    p_death = deaths_per_100k / 100_000 / prevalence_percent
    return p_death


def estimate_dur_condition(prevalence_rate, incidence_rate):

    if incidence <= 0:
        raise ValueError("Incidence must be greater than 0.")

    return prevalence_rate / incidence_rate
```

```python
# Type 2 Diabetes

# Input values
deaths_per_100k = 55.81
prevalence_percent = 0.0225
prevalence = 2171.05
incidence = 177.33

# Estimate p_death and dur_condition
p_death = estimate_p_death(deaths_per_100k, prevalence_percent)
dur_condition = estimate_dur_condition(prevalence, incidence)

print(f"Estimated p_death (T2D, Eswatini 2007): {p_death:.4f} per year")
print(f"Estimated duration of T2D: {dur_condition:.2f} years")
```
